```python
import pandas as pd
import numpy as np


df =  pd.read_csv("/content/task01.csv")
df =  np.array(df)
np.random.shuffle(df)

x = df[:, 1]
y = df[:, 2]
def loss_func(y_true,y_pred):
  n=len(y_true)
  ans=1/n*sum(((y_true-y_pred)*(y_true-y_pred)))
  return ans
trainSize=int(0.7*len(x))
X_Train = x[:trainSize]
Y_Train = y[:trainSize]

testSize = int(0.15*len(x))
X_Test = x[trainSize:trainSize+testSize]
Y_Test = y[trainSize:trainSize+testSize]
        
valSize = int(0.15*len(x))
X_Val = x[trainSize+testSize:trainSize+testSize+valSize+1]
Y_Val = y[trainSize+testSize:trainSize+testSize+valSize+1]

lr =0.0000000001
epochs=1200
m = np.random.uniform(np.mean(Y_Train)/np.mean(X_Train),np.mean(Y_Train)/np.mean(X_Train)+1000)
c=np.random.uniform(np.mean(Y_Train),np.mean(Y_Train)+1000)
z=0
n = len(x)
train_loss=[]
val_loss=[]
for i in range(epochs):

  Ygrad_train=(m*X_Train)+c
  Ygrad_val = (m*X_Val) + c,
  mt=loss_func( Y_Train,Ygrad_train)
  mv=loss_func( Y_Val,Ygrad_val)
  train_loss.append(mt)
  val_loss.append(mv)
  derM=1/n*np.sum((Ygrad_train-Y_Train)*X_Train)
  derC=1/n*np.sum((Ygrad_train-Y_Train))
  m-=lr*derM
  c-=lr*derC


#print(trainX,trainY,testX,testX,testY,valX,valY)
```
